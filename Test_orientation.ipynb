```python
import numpy as np
import matplotlib.pyplot as plt

cube = np.zeros((3,3,3))

cube[:,0,0] = 1
cube[0,:,0] = 2
cube[0,0,:] = 3
cube[0,0,0] = 5
```

```python
def plot(cube, ax):

    xs = []
    ys = []
    zs = []
    color = []
    for x in range(3):
        for y in range(3):
            for z in range(3):
                if cube[x,y,z] != 0:
                    xs.append(x)
                    ys.append(y)
                    zs.append(z)
                    if cube[x,y,z] == 1:
                        color.append('blue')
                    elif cube[x,y,z] == 2:
                        color.append('red')   
                    elif cube[x,y,z] == 3:
                        color.append('green')
                    elif cube[x,y,z] == 5:
                        color.append('black')

    ax.scatter(xs, ys, zs, c=color, cmap='viridis', s=100)

initial_cube = cube.copy()


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plot(cube, ax)
```

# Sum over X

```python
obs_x = np.sum(initial_cube, axis=0)

cube_x = np.zeros((3,3,3))

for i in range(3):
    for j in range(3):
        cube_x[i,j,:] = initial_cube[:,i,j]

im_ratio = obs_x.shape[0]/obs_x.shape[1]

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
im = axs[0].imshow(obs_x.T, cmap='tab20', origin='lower')
plt.colorbar(im, fraction=0.047*im_ratio)
im = axs[1].imshow(np.sum(cube_x, axis=-1).T, cmap='tab20', origin='lower')
plt.colorbar(im, fraction=0.047*im_ratio)
im = axs[2].imshow(cube_x[:,:,0].T, cmap='tab20', origin='lower')
plt.colorbar(im, fraction=0.047*im_ratio)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(121, projection='3d')
plot(initial_cube, ax)
ax = fig.add_subplot(122, projection='3d')
plot(cube_x, ax)

print(obs_x[:,0])
print(obs_x[0,:])
```
