```python
import numpy as np
import matplotlib.pyplot as plt

cube = np.zeros((3,3,3))

cube[:,0,0] = 1
cube[0,:,0] = 2
cube[0,0,:] = 3
cube[0,0,0] = 5
```

```python
def plot(cube):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    xs = []
    ys = []
    zs = []
    color = []
    for x in range(3):
        for y in range(3):
            for z in range(3):
                if cube[x,y,z] != 0:
                    xs.append(x)
                    ys.append(y)
                    zs.append(z)
                    if cube[x,y,z] == 1:
                        color.append('red')
                    elif cube[x,y,z] == 2:
                        color.append('green')   
                    elif cube[x,y,z] == 3:
                        color.append('blue')
                    elif cube[x,y,z] == 5:
                        color.append('black')

    ax.scatter(xs, ys, zs, c=color, cmap='viridis', s=100)

    plt.show()

plot(cube)
```

```python
plt.imshow(np.sum(cube, axis=0), cmap='Set1')
plt.colorbar()  
plt.show()
```

```python
cube = np.flip(cube, axis=1)
cube = cube.swapaxes(0,2)

plot(cube)
```
